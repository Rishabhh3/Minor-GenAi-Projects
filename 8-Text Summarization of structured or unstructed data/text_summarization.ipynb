```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq
api_key = os.getenv("GROQ_API_KEY")
llm = ChatGroq(api_key = api_key,model ="llama-3.3-70b-versatile")
llm
```

```python
from langchain_classic.schema import(
    AIMessage,
    HumanMessage, SystemMessage
)
```

```python
speech = """ In today’s fast-paced world, the ability to process and understand information efficiently has become more important than ever. Every day, we are exposed to an overwhelming amount of data—through books, articles, social media, and conversations. However, simply reading or listening is not enough. What truly matters is how well we can comprehend, organize, and retain this information. This is where the skill of summarization plays a vital role.

Summarization is not just about shortening a piece of text; it is about identifying the core ideas and presenting them in a clear and concise manner. It requires critical thinking, attention to detail, and the ability to distinguish between essential and non-essential information. By practicing summarization, we train our minds to focus on what truly matters, filtering out distractions and unnecessary details.

One of the key benefits of summarization is that it improves comprehension. When we attempt to summarize a text, we are forced to engage deeply with the material. We ask ourselves questions such as: What is the main idea? What are the supporting arguments? Which details are crucial, and which can be omitted? This active engagement leads to a better understanding of the content.

Moreover, summarization enhances memory retention. Studies have shown that when we process information and restate it in our own words, we are more likely to remember it. This is because summarization involves both understanding and reconstruction, which strengthens neural connections in the brain.

In academic and professional settings, summarization is an invaluable skill. Students use it to condense large chapters into manageable notes, while professionals rely on it to present reports, proposals, and briefings efficiently. In both cases, the ability to communicate key ideas clearly and concisely can make a significant difference.

However, mastering summarization is not an overnight process. It requires consistent practice. One effective approach is to start with shorter texts and gradually move to more complex material. Another useful technique is to compare your summary with the original text to identify what you may have missed or misunderstood.

In conclusion, summarization is a powerful skill that enhances comprehension, improves memory, and enables effective communication. In a world overflowing with information, it allows us to stay focused, organized, and informed. By dedicating time to practice and refine this skill, we not only become better learners but also more efficient thinkers and communicators.
 """
```

```python
chat_message = [
    SystemMessage(content="You are an expert in summarizing speeches"),
    HumanMessage(content=f"Please provide a short and concise summary of following speech:\n Text:{speech}")
]
```

```python
llm.get_num_tokens(speech)
```

```python
reponse = llm.invoke(chat_message) # response = AIResponse
print(reponse.content)
```

# Prompt Template Text Summarization

```python

from langchain_core.prompts import PromptTemplate

generic_prompt = """
    Write a summary of following speech:
    speech:{speech}
    Translate the precise summary to:{language}
"""

prompt = PromptTemplate(
    input_variables=['speech','language'],
    template = generic_prompt
)
prompt
```

```python
complete_prompt = prompt.format(speech = speech, language="Hindi")
llm.invoke(complete_prompt)
```

```python
llm.get_num_tokens(complete_prompt)
# increased bcoz along with speech other things are passed on
```

```python
chain = prompt | llm
summary = chain.invoke({'speech':speech,'language':'Hindi'})
print(summary.content)
```
